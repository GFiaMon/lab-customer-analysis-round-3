```python
sns.histplot(df['response'], kde=False)
plt.title('Distribution of Response')
plt.xlabel('Response')
plt.ylabel('Frequency')
plt.show()

# Distribution of passenger ages
# sns.histplot(titanic['age'].dropna(), kde=True)
# plt.title('Distribution of Passenger Ages')
# plt.xlabel('Age')
# plt.ylabel('Frequency')
# plt.show()
```

- Show a plot of the response rate by the sales channel.


```python
# df['response'].value_counts() / len(df['response'].dropna())
```

```python
# df['response'].value_counts(normalize=True)
```

```python
# df['response_n'] = df['response'].value_counts(normalize=True)

response_rate_by_channel = df.groupby('response')['sales_channel'].value_counts(normalize=True).unstack()
response_rate_by_channel
```

```python
sns.barplot(x=response_rate_by_channel.columns, y=response_rate_by_channel.loc['Yes'])
```

```python
# Convert 'response' to a binary indicator if it isn't already (assuming responses are 'Yes'/'No')
df['response_binary'] = df['response'].apply(lambda x: 1 if x == 'Yes' else 0)

# Calculate response rate by sales channel
response_rate_by_channel = df.groupby('sales_channel')['response_binary'].mean().reset_index()
display(response_rate_by_channel)
```

```python

sns.barplot(x='sales_channel', y='response_binary', data=response_rate_by_channel)
plt.title('Response Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate')
plt.show()
```

- Show a plot of the response rate by the total claim amount.


```python
response_rate_by_claim = df.groupby('response')['total_claim_amount'].value_counts(normalize=True).unstack()
response_rate_by_claim
```

```python
#Show a plot of the response rate by the total claim amount.
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['response'], y=df['total_claim_amount'], palette='pastel')
plt.xlabel('Response')
plt.ylabel('Total Claim Amount')
plt.title('Response Rate by Total Claim Amount')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
